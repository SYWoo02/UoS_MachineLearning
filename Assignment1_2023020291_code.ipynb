```python
import numpy as np
import matplotlib.pyplot as plt

# 랜덤 데이터 생성
x = np.random.uniform(-10,10,30)
y = np.random.uniform(-10,10,30)

plt.figure(figsize=(12, 12))

# 임의의 직선을 위한 데이터 생성
# 기울기 : slope
# y절편 : slope
slope = []
intercept = []
for i in range(0,4):
  slope.append(np.random.uniform(-5, 5))
  intercept.append(np.random.uniform(-10, 10))

x_above = []
x_below = []
y_above = []
y_below = []
for i in range(0,4):

  x_above.clear()
  x_below.clear()
  y_above.clear()
  y_below.clear()

  for j in range(0,30):
    if y[j] > slope[i] * x[j] + intercept[i]:
      # 임의의 직선 위의 점들
      x_above.append(x[j])
      y_above.append(y[j])
    else:
      # 임의의 직선 아래의 점들
      x_below.append(x[j])
      y_below.append(y[j])

  # 임의의 직선
  yline = slope[i] * x + intercept[i]

  # 2x2 서브플롯 생성
  plt.subplot(2, 2, i+1)

  # 임의의 직선 그리기
  plt.plot(x, yline, color = 'red')
  # 산점도 그리기 (임의의 직선 위 : 초록색의 별 모양)
  plt.scatter(x_above, y_above, color = 'green', marker='*')
  # 산점도 그리기 (임의의 직선 아래 : 파란색의 세모 모양)
  plt.scatter(x_below, y_below, color = 'blue', marker='^')

  plt.xlabel('x')
  plt.ylabel('y')

  plt.xlim(-10,10)
  plt.ylim(-10,10)
```
